```python
#importing libs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Task 1 : Load and explore the dataset


try:
    # load the dataset
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
    data  = pd.read_csv(url)

    #displaying first 5 rows of the dataset
    print("The first 5 rows of the dataset :")
    print(data.head())

    #checking data types and missing values
    print("\nDataset info :")
    print(data.info())

    print("\nChecking for missing vals :")
    print(data.isnull().sum())

except FileNotFoundError:
    print("Error : File not found, please check the file path again.")
except Exception as e :
    print(f"An unknown error occured : {e}")


#Task 2 : Basic Data Analysis
try:
    #basic staistics
    print("\nBasic statisics :")
    print(data.describe())

    #grouping mean petal length per species
    grouped_data = data.groupby("species")["petal_length"].mean()
    print("\nMean petal length per species :")
    print(grouped_data)


    #observations
    print("\nObservations :")
    print("1. Versicolor species has the highest average petal length.")
    print("2. Setosa species has the smallest average petal length .")
except Exception as e:
    print(f"An error occured during analysis : {e}")


    #Task 3 . Data Viz

try:
    #line chart: trend of petal_length over rows
    plt.figure(figsize=(8,15))
    plt.plot(data.index, data["petal_length"], label="Petal Length", color="green")
    plt.title("Petal length trend")
    plt.xlabel("Index")
    plt.ylabel("Petal length")
    plt.legend()
    plt.show()


    #bar chart
    plt.figure(figsize=(8,15))
    grouped_data.plot(kind = "bar", color = ["blue", "orange", "green"])
    plt.title("Average petal length per species")
    plt.xlabel("Species")
    plt.ylabel("Petal length")
    plt.show()




    #histogram distro for petal length

    plt.figure(figsize=(8,15))
    plt.hist(data["Petal_length"], bins = 20, color = "blue", edgecolor = "black")
    plt.title("Distribution of petal lengths")
    plt.xlabel("Petal length")
    plt.ylabel("Frequency")
    plt.show()


    #scatter plot : sepal length vs petal length
    plt.figure(figsize=(8,15))
    sns.scatterplot(x="sepal_length", y="petal_length", hue = "species", data=data, pallete = "viridis")
    plt.title("Sepal length vs Petal length")
    plt.xlabel("Sepal length")
    plt.ylabel("Petal length")
    plt.legend(title="Species")
    plt.show()

except Exception as e:
    print(f"An error occured during visualization : {e}")
```
